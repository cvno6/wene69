```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from matplotlib.patches import Ellipse
from scipy.stats import multivariate_normal

# Generate synthetic dataset
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=1.0, random_state=42)

# Initialization
n_samples, n_features = X.shape
k = 3  # Number of clusters
np.random.seed(42)
weights = np.ones(k) / k
means = X[np.random.choice(n_samples, k, replace=False)]
covariances = np.array([np.cov(X.T) for _ in range(k)])
log_likelihoods = []

def draw_ellipses(means, covariances, ax, color='red'):
    for pos, covar in zip(means, covariances):
        if covar.shape == (2, 2):
            U, s, Vt = np.linalg.svd(covar)
            angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
            width, height = 2 * np.sqrt(s)
        else:
            angle = 0
            width, height = 2 * np.sqrt(covariances)
        for nsig in range(1, 3):
            ellipse = Ellipse(xy=pos, width=nsig*width, height=nsig*height,
                             angle=angle, edgecolor=color, fc='None', lw=2, alpha=0.5)
            ax.add_patch(ellipse)

# EM iterations
max_iter = 20
for iteration in range(max_iter):
    # ----- E-Step -----
    responsibilities = np.zeros((n_samples, k))
    for i in range(k):
        rv = multivariate_normal(mean=means[i], cov=covariances[i])
        responsibilities[:, i] = weights[i] * rv.pdf(X)
    responsibilities /= responsibilities.sum(axis=1, keepdims=True)

    # ----- M-Step -----
    Nk = responsibilities.sum(axis=0)
    weights = Nk / n_samples
    means = (responsibilities.T @ X) / Nk[:, np.newaxis]
    covariances = np.zeros((k, n_features, n_features))
    for i in range(k):
        diff = X - means[i]
        covariances[i] = (responsibilities[:, i][:, np.newaxis] * diff).T @ diff / Nk[i]

    # ----- Plotting E-step and M-step side-by-side -----
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # E-step plot: responsibilities (argmax)
    axes[0].scatter(X[:, 0], X[:, 1], c=responsibilities.argmax(axis=1), cmap='viridis', s=40)
    axes[0].set_title(f"E-step - Responsibilities (Iter {iteration+1})")
    axes[0].set_xlabel("Feature 1")
    axes[0].set_ylabel("Feature 2")
    axes[0].grid(True)

    # M-step plot: updated means & covariances
    axes[1].scatter(X[:, 0], X[:, 1], c=responsibilities.argmax(axis=1), cmap='viridis', s=40)
    draw_ellipses(means, covariances, axes[1])
    axes[1].set_title(f"M-step - Updated Gaussians (Iter {iteration+1})")
    axes[1].set_xlabel("Feature 1")
    axes[1].set_ylabel("Feature 2")
    axes[1].grid(True)

    plt.tight_layout()
    plt.show()

    # ----- Log-Likelihood -----
    log_likelihood = np.sum(np.log(np.sum([
        weights[i] * multivariate_normal(mean=means[i], cov=covariances[i]).pdf(X)
        for i in range(k)
    ], axis=0)))
    log_likelihoods.append(log_likelihood)

# Final: Log-likelihood convergence plot
plt.figure(figsize=(8, 4))
plt.plot(range(1, max_iter + 1), log_likelihoods, marker='o', linestyle='--')
plt.title("Log-Likelihood Convergence")
plt.xlabel("Iteration")
plt.ylabel("Log-Likelihood")
plt.grid(True)
plt.show()
```
