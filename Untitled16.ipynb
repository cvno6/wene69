```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.linspace(0, 10, 50)

# quadratic: y = 2x^2 + 3x + 5
true_y = 2 * x**2 + 3 * x + 5

noise = np.random.normal(0, 20, size=x.shape)
y_noisy = true_y + noise

degree = 2
coefficients = np.polyfit(x, y_noisy, degree)

polynomial = np.poly1d(coefficients)

x_smooth = np.linspace(min(x), max(x), 500)
y_smooth = polynomial(x_smooth)

plt.scatter(x, y_noisy, color='red', label='Noisy Data Points')
plt.plot(x_smooth, y_smooth, color='blue', label=f'Polynomial Fit (degree={degree})')
plt.plot(x, true_y, color='green', linestyle='--', label='True Underlying Curve')
plt.title('Polynomial Curve Fitting with Noise')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
```
