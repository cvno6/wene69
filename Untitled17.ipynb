```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target

petal_length = X[:, 2]
petal_width = X[:, 3]

degree = 3

coefficients = np.polyfit(petal_length, petal_width, degree)

polynomial = np.poly1d(coefficients)

x_smooth = np.linspace(min(petal_length), max(petal_length), 500)
y_smooth = polynomial(x_smooth)

plt.scatter(petal_length, petal_width, color='red', label='Iris Data Points')
plt.plot(x_smooth, y_smooth, color='blue', label=f'Polynomial Fit (degree={degree})')
plt.title('Polynomial Curve Fitting on Iris Dataset (Petal Length vs Width)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()

print("Fitted Polynomial Equation:")
print(polynomial)
```
